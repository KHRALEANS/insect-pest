```python
import matplotlib.pyplot as plt
import numpy as np


plt.rcParams['font.family'] = 'SimHei'
# plt.rcParams['font.sans-serif'] = ['SimHei', 'Noto Sans CJK SC', 'Noto Serif CJK SC', 'Noto Sans Mono CJK SC']


plt.clf()
fig, ax = plt.subplots(figsize=(10, 5),dpi=500)

insect_name = ['虫虫1','虫虫2']
y_pos = np.arange(len(insect_name))

ax.barh(y_pos, [3,4], align='center')
# ax.set_yticks(y_pos, labels=insect_name,rotation=0 , fontsize=8)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('数量')
ax.set_title('各类虫子数量')

plt.show()
```
